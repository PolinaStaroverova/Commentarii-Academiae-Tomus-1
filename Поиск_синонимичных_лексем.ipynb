```python
df = pd.DataFrame(columns=['Name of text', 'ergo', 'igitur'])
for path_to_file in lst_of_pathes:
  new_item = []
  match_ergo = []
  match_igitur = []
  text = ''
  with open(path_to_file, 'r') as open_file:
    text = open_file.read().lower()
  text = re.sub(r'[^\w\s]','', text)
  print(path_to_file)
  text[:100]

  match_ergo = re.findall(r'ergo', text)
  match_igitur = re.findall(r'igitur', text)

  new_item.append(path_to_file)
  new_item.append(len(match_ergo))
  new_item.append(len(match_igitur))

  df.loc[ len(df)] = new_item
```
